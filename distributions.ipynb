Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt  #for making graphs and plots
#pyplot module inside it is used to quickly create plots, charts, and graphs.
from scipy.stats import norm   #scipy is a scientific computing library , norm means normal distribution.
#scipy.stats is a module inside scipy that provides statistical functions, like probability distributions.
```

```python
mean=0   #mean = 0 means the curve is centered at 0
std_dev=1    #std_dev=1 "How spread out is the data?" , tail of graph
x = np.linspace(-4,4,1000)  #np.linspace(start, stop, num)
#represents the points of x-axis
```

```python
pdf=norm.pdf(x,mean,std_dev)   #Probability Density Function (PDF) of a normal distribution
#PDF tells you how dense the probability is at each point
```

```python
#draws a line plot
plt.plot(x, pdf, color='green', label="Normal distribution")
plt.title("Normal Distribution (mean=0, std=1)")
plt.xlabel("x")  #horizontal axis
plt.ylabel("Probability Density")  #vertical axis
plt.grid(True) #turns on grid lines
plt.legend()  #this shows label on the graph
plt.show()
```

Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom #for binomial distribution
```

```python
n=10 #number of trials
p=0.5  #probabilty of success
x=np.arange(0, n+1)  #numpy array from 0 to 10
```

```python
pmf= binom.pmf(x, n, p) #Probability Mass Function (PMF)
```

```python
plt.bar(x, pmf, color="skyblue")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of successes")
plt.ylabel("Probability Distribution")
plt.grid(axis='y')
plt.show()
```

Poisson Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson #for poisson distribution
```

```python
lambda_val=3 #average rate of occurrances
x= np.arange(0,15) #array from 0 to 14
```

```python
pmf=poisson.pmf(x, mu=lambda_val) #mu represents mean rate
```

```python
plt.bar(x, pmf, color="orange")
plt.title("Poisson Distribution (lambda=3)")
plt.xlabel("Number of events")
plt.ylabel("Probability Distribution")
plt.grid(axis='y')
plt.show()
```
