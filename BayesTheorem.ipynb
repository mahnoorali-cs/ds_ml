```python
def bayes_theorem(prior, likelihood, false_positive, prior_not=None):

    if prior_not is None:
        prior_not = 1 - prior

    evidence = (likelihood * prior) + (false_positive * prior_not)
    posterior = (likelihood * prior) / evidence

    return posterior
```

```python
def get_medical_test_data():
    """
    Returns example medical test data.
    """
    data = {
        "prior": 0.01,              # P(H): Probability of having the disease
        "likelihood": 0.99,         # P(E|H): Probability test is positive given disease
        "false_positive": 0.01      # P(E|Â¬H): Probability test is positive given no disease
    }
    return data
```

```python
def main():
    # Load example data
    data = get_medical_test_data()

    # Apply Bayes' Theorem
    posterior = bayes_theorem(
        prior=data["prior"],
        likelihood=data["likelihood"],
        false_positive=data["false_positive"]
    )

    # Output the result
    print(f"Probability of having the disease given a positive test result: {posterior:.4f}")


if __name__ == "__main__":
    main()
```
