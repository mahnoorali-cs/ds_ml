Reinforcement Learning

```python
import random
import matplotlib.pyplot as plt
```

```python
class SimpleEnvironment:
    def __init__(self):
        self.state = 0  # Start state
    def step(self, action):
        reward = 0
        if action == 1:  # Move right
            self.state += 1
            reward = 1
        elif action == 0:  # Move left
            self.state -= 1
            reward = -1
        return self.state, reward
```

```python
q_table = {0: [0, 0]}
env = SimpleEnvironment()
states_visited = []
rewards_collected = []
```

```python
for episode in range(10):
    state = env.state
    action = random.choice([0, 1])  # Choose an action randomly (move left or right)
    next_state, reward = env.step(action)
    # Check if the state exists in the Q-table, if not, add it
    if state not in q_table:
        q_table[state] = [0, 0]  # Initialize with 0 values for both actions
    q_table[state][action] = reward  # Update Q-table with the received reward
    # Store the state and reward for plotting
    states_visited.append(state)
    rewards_collected.append(reward)
    print(f"Episode {episode+1} - State: {state}, Action: {action}, Reward: {reward}")
print("\nFinal Q-Table:", q_table)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(states_visited, rewards_collected, marker='o', linestyle='-', color='b', label="Reward")
plt.xlabel('State')
plt.ylabel('Reward')
plt.title('Reinforcement Learning: State vs Reward over Episodes')
plt.legend()
plt.grid(True)
plt.show()
```
